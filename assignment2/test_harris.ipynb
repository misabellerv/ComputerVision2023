```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from harris_detector import Harris
```

```python
building1 = cv2.imread('C:/Users/Isabelle/Desktop/computervision/Listas_e_dados/Listas_e_dados/CV_Lista02/CV_Lista02_dados/building1.jpg')
building2 = cv2.imread('C:/Users/Isabelle/Desktop/computervision/Listas_e_dados/Listas_e_dados/CV_Lista02/CV_Lista02_dados/building2.jpg')

k = 0.04
wd = 5
threshold = 0.027

harris_response_building1, keypoints_building1 = Harris(building1, k,wd, threshold)
harris_response_building2, keypoints_building2 = Harris(building2, k,wd, threshold)

for kp in keypoints_building1:
    cv2.circle(building1, (kp[1], kp[0]), 1, (0, 255, 0), -1)

for kp in keypoints_building2:
    cv2.circle(building2, (kp[1], kp[0]), 1, (0, 255, 0), -1)

plt.figure(figsize=(10, 3))
plt.subplot(1, 2, 1)
plt.imshow(building1)
plt.subplot(1, 2, 2)
plt.imshow(building2)
plt.suptitle(f'Harris detector (Wd = {wd}, threshold = {threshold})', fontsize=12)
```
